```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn3
```

```python
v = venn2(subsets = [set([1,2]), set(([2,3,4,5]))], set_labels = ('a', 'b'))

plt.title(" Venn diagram")
plt.show()
```

```python
v2 = venn3(subsets=(1,1,0,1,0,0,0), set_labels=('a', 'b', 'c'))
v2.get_label_by_id('100').set_text('First')
v2.get_label_by_id('010').set_text('Second')
v2.get_label_by_id('001').set_text('Third')
plt.title("Not a Venn diagram")
plt.show()
```
